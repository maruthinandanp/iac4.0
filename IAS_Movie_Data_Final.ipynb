## Converting Budget and Gross into  a numeric column:**

```python
def num_convert(amount):
    if str(amount)[:3] == 'GBP':
        gbp =  int(amount[3:].replace(',',''))
        return gbp*1.3
    elif str(amount)[:2] == 'EU':
        eu = int(amount[2:].replace(',',''))
        return eu*1.1
    elif str(amount)[:3] == 'CAD':
        cad =  int(amount[3:].replace(',',''))
        return cad*0.75
    elif len(str(amount)) > 20:
        return np.nan
    else:
        return amount
```

```python
data['Gross'] = data['Gross'].apply(num_convert)
data['Gross']= data['Gross'].apply(pd.to_numeric, errors = 'ceorce')
```

```python
data['Budget_x'] = data['Budget_x'].apply(num_convert)
data['Budget_x']= data['Budget_x'].apply(pd.to_numeric, errors = 'ceorce')
```
